```matlab
f = @(t,x) 2*x;
t = linspace(1,5,300);
[t,x] = ode45(f,t,3);
clf
plot(t,x)
xlabel('t'), ylabel('x')
```

```matlab
semilogy(t,x)
xlabel('t'), ylabel('x')
```

```matlab
f = @(t,x) 2*t*x;
t = linspace(0,5,300);
[t,x] = ode45(f,t,1);
clf
semilogy(t,x)
xlabel('t'), ylabel('x')
```
