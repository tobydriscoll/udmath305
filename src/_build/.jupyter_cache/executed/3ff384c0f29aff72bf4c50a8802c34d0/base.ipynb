```matlab
a = 1.25;
for C = [-1 -0.5 -0.25 0 0.25 0.5 1]
    fplot(@(t) C*exp(a*t),[1,4])
    hold on
end

plot(2,5,'k.','markersize',16)
fplot(@(t) 5*exp(a*(t-2)),[1,4],'k','linew',2)
ylim([-100 100])
```

```matlab
f = @(t,x) 2*x;
t = linspace(1,5,300);
[t,x] = ode45(f,t,3);
clf
plot(t,x)
xlabel('t'), ylabel('x')
```

```matlab
semilogy(t,x)
xlabel('t'), ylabel('x')
```

```matlab
f = @(t,x) 2*t*x;
t = linspace(0,5,300);
[t,x] = ode45(f,t,1);
clf
semilogy(t,x)
xlabel('t'), ylabel('x')
```

```matlab
f = @(t,x) x^2;
t = linspace(0,4,300);
[t,x] = ode45(f,t,0.5);
semilogy(t,x)
xlabel('t'), ylabel('x')
```
